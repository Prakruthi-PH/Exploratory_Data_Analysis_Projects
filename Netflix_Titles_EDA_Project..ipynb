# **Exploratory Data Analysis of netflix_titles**


**Problem Statement**

The objective of this analysis is to examine the Netflix content dataset to understand the distribution of Movies and TV Shows, identify growth trends over time, and evaluate key factors such as country contribution, genre popularity, ratings, and duration patterns that influence content strategy. This study aims to uncover meaningful insights through data cleaning, transformation, and exploratory data analysis to assess year-wise and month-wise content expansion and audience targeting trends. The findings will support data-driven decision-making by highlighting content performance patterns, geographic distribution, and strategic opportunities for platform growth.


**Objectives**

To perform Exploratory Data Analysis (EDA) on the Netflix dataset.

To clean and preprocess the dataset by handling missing values, extracting date components (Year, Month, Day), and transforming categorical variables.

To analyze the proportion of Movies vs TV Shows.

To identify top contributing countries in content production.

To determine the most common genres available on Netflix.

To examine the distribution of content ratings across the platform.

To analyze yearly and monthly trends in content addition.

To identify patterns, correlations, and anomalies within the dataset.

To generate meaningful business insights that could support content strategy decisions.

**Data Descriptions**



**show_id** – Unique ID given to each movie or TV show.

**type** – Tells whether the content is a Movie or a TV Show.

**title** – Name of the movie or TV show.

**director** – Name of the director of the content.

**cast** – Names of actors and actresses in the content.

**country** – Country where the content was produced.

**date_added** – Date when the content was added to Netflix.

**release_year** – Year when the content was originally released.

**rating** – Age classification given to the content (like TV-MA, PG, etc.).

**duration** – Length of the content (minutes for movies, seasons for TV shows).

**listed_in** – Genre or category of the content.

**description** – Short summary of the movie or TV show.

# Handling data and time

```python
dframe1.columns
```

```python
# Date column is in object so need to convert to standard date
dframe1['date_added'].dtype
```

```python
# Converting date column to standard format
dframe1['date_added'] = pd.to_datetime(dframe['date_added'].str.strip(),errors='coerce')
```

```python
# Creating the new date column
dframe1['Year'] = dframe1['date_added'].dt.year
```

```python
# Creating the new month column
dframe1['Month'] = dframe1['date_added'].dt.month
```

```python
dframe1.columns
```

```python
# Dropped date added column
dframe1.drop(columns=['date_added'], inplace=True)
```

```python
# Renaming column Year to year_added
dframe1.rename(columns={'Year': 'year_added'},inplace=True)
```

```python
# Renaming column  Month to month_added
dframe1.rename(columns={'Month': 'month_added'},inplace=True)
```

```python
dframe1.columns
```

```python
dframe1['release_year'].dtypes
```

```python
# Splitting duration column into duration_type and duration_value
# Extract the string part from duration column
dframe1['duration_type'] = dframe1['duration'].str.extract('([a-zA-Z]+)')
```

```python
# Extracting numeric part from duration column
dframe1['duration_value'] = dframe1['duration'].str.extract('(\d+)')
```

```python
dframe1['duration_value'].dtypes
```

```python
# Changing the datatype of duration_value from object to  int64
dframe1['duration_value'] = dframe1['duration_value'].astype('Int64')
```

```python
# Converted to int
dframe1['duration_value'].dtypes
```

```python
# Dropped duration column
dframe1.drop(columns=['duration'], inplace=True)
```

```python
dframe1['type'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Movie vs TV Shows')
plt.ylabel('')
plt.show()
```

 **Movies dominate the content library** — About **69.6%** of the titles are movies, showing a strong focus on film content compared to series.

 **TV shows form a smaller but significant portion** — With **30.4%**, TV shows still represent nearly one-third of the catalog, indicating meaningful but secondary investment.

 **Content strategy appears movie-centric** — The large gap (almost 40 percentage points) suggests the platform prioritizes movies over episodic storytelling, possibly to offer quick-watch entertainment or a broader film variety.



```python
movies = dframe1[dframe1['duration_type'] == 'min']
plt.scatter(movies['release_year'], movies['duration_value'])
plt.title('Release Year vs Movie Duration')
plt.xlabel('Release Year')
plt.ylabel('Duration (Minutes)')
plt.show()
```



**Most movies are released after 2000**
   The highest concentration of data points appears from 2000 onward, showing that most titles in the dataset are relatively recent.

**Typical movie duration ranges between 80–150 minutes**
   The majority of films cluster within this range, suggesting it is the standard runtime regardless of release year.

**No strong relationship between release year and duration**
   Movie lengths vary widely across all years, indicating that duration has remained fairly consistent over time without a clear upward or downward trend.


```python
yearly_content = dframe1['year_added'].value_counts().sort_index()
plt.plot(yearly_content.index, yearly_content.values)
plt.title('Content Added Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.show()
```



**Rapid growth after 2015**
   Content additions were very low before 2015, then increased sharply from 2016 onward, showing a major expansion phase.

**Peak around 2019**
   The highest number of titles was added around 2019, marking the platform’s most aggressive content growth period.

**Recent decline after the peak**
   After 2019, the number of titles added each year begins to drop slightly, suggesting slower expansion or a shift in content strategy.


```python
plt.figure(figsize=(5,3))
dframe1['type'].value_counts().plot(kind='bar')
plt.title('Movies vs TV Shows')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```



**Movies significantly outnumber TV shows** — The number of movies is more than double the number of TV shows, showing a strong content imbalance.

 **Platform focuses more on film content** — The large gap suggests a strategic emphasis on movies rather than episodic series.

**TV shows still represent a substantial share** — Although fewer than movies, TV shows still make up a notable portion of the catalog, indicating continued investment in series content.


```python
dframe1['listed_in'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Genres')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

**Drama-based international content leads strongly**
   “Dramas, International Movies” is the most common genre, showing a major emphasis on global storytelling and dramatic themes.

**Documentaries and stand-up comedy are highly popular**
   Both appear near the top, indicating strong demand for real-life content and comedy specials alongside traditional movie genres.

**Family and kids’ content is present but not dominant**
   Categories like “Kids’ TV” and “Children & Family Movies” appear in the top 10 but with lower counts, suggesting the platform prioritizes adult and general audience entertainment more than children-focused content.


```python
dframe1['rating'].value_counts().plot(kind='bar')
plt.title('Content Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

**Mature content dominates the catalog**
   The **TV-MA** category has the highest count by a large margin, indicating the platform mainly targets adult audiences with mature themes.

 **Teen and general audience content is the second largest segment**
   Ratings like **TV-14**, **TV-PG**, and **R** also have substantial numbers, showing strong availability of content suitable for teenagers and general adult viewing.

 **Very limited children and niche ratings**
   Categories such as **TV-Y**, **TV-Y7**, **G**, and specialized or uncommon ratings have much smaller counts, suggesting the platform has comparatively less focus on young children’s programming or rare rating categories.


**Summary**


The dataset shows a higher proportion of Movies compared to TV Shows, indicating a stronger focus on movie-based content.

Content addition significantly increased after 2015, reflecting rapid platform expansion during that period.

A small number of countries contribute the majority of content, with dominant production coming from leading entertainment markets.

Drama, International Movies, and Comedies emerge as the most frequent genres in the catalog.

Most content falls under mature audience ratings (such as TV-MA), suggesting a strong focus on adult viewership.


**Conclusion**

Netflix has strategically expanded its content library over recent years to strengthen global market presence.

The platform demonstrates a movie-heavy catalog while gradually increasing TV Show offerings.

Genre distribution indicates a clear preference for high-engagement categories like Drama and Comedy.

The concentration of content from select countries highlights geographic production dominance.

Audience targeting is primarily oriented toward mature viewers based on rating distribution.
